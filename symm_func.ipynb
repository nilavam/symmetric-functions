```python
#L1 and L2 must be permutations of each other
def standard_perm(L1, L2):
    perm = Word(L2).standard_permutation() / Word(L1).standard_permutation()
    # L1[perm[i]] = L2[i]
    #note this is a right action on L1
    return perm.cycle_string()
standard_perm([1,2,3, 4], [3,4,1,2])
```

```python
#p is a sylbolic partition
def irrep_Sym_symb(p):
    n = len(p)
    dim = 1
    for i in range(n):
        for j in range(i+1,n):
            dim = dim * ((p[i]-p[j])/(j-i) + 1)
    return dim

vars = []
for i in range(4):
    vars.append(var('x{}'.format(i)))
vars
a = irrep_Sym_symb(vars) #schur_{x0,x1,x2,...}(1,1,...,1)
print(a)
a.subs(x0 == 2, x1 == 1, x2 == 1, x3 == 0)
```

```python
def conjugate(p):
    x = []
    l = len(p)
    for i in range(1,p[0]+1):
        s = 0
        for j in range(l):
            if (p[j] >= i):
                s += 1
            else:
                break
        x.append(s)
    return x
conjugate([4,2,1])
```

```python
def n(p):
    l = len(p)
    s = 0
    for i in range(l):
        s += i * p[i]
    return s
```

```python
def content_poly(p):
    x = var('x')
    l = len(p)
    poly = 1
    for i in range(l):
        for j in range(p[i]):
            poly *= (x+j-i)
    return poly
content_poly([2,1]).subs(x==2)
```

```python
#schur_lambda(1,1,1,....,1), d many 1's
def schur_at_one(p, d = var('d')):
    l = len(p)
    s = content_poly(p).subs(x==d)
    q = conjugate(p)
    for i in range(l):
        for j in range(p[i]):
            #since i,j are indexed from 0, replace i,j by i+1,j+1 respectively
            arm = p[i] - j - 1
            leg = q[j] - i - 1
            s = s / (arm + leg + 1)
    return s
schur_at_one([2,1,1])
```

```python
def findStab2_multiplicity(mult):
    prefix = [0]
    gen = []
    for i in mult:
        prefix.append(prefix[-1]+i)
        gen.append(tuple([j+prefix[-2]+1 for j in range(i)]))
        if(i > 1):
            gen.append(tuple([j+prefix[-2]+1 for j in range(2)]))
    G = PermutationGroup(gen)
    return G

def findStab2(l):
    x = l.copy()
    l.sort()
    mult = []
    temp = l[0]
    temp_freq = 0
    S = SymmetricGroup(len(x))
    for i in l:
        if(i != temp):
            temp = i
            mult.append(temp_freq)
            temp_freq = 1
        else:
            temp_freq = temp_freq + 1
    mult.append(temp_freq)
    G = findStab2_multiplicity(mult)
    sigma = (Word(l).standard_permutation() / Word(x).standard_permutation()).cycle_string()
    s = S(sigma)
    H = [s^(-1) * S(i) * s for i in G.list()]
    print(s)
    print(G.list())
    print(H)
    
findStab2([1,4,2,4,1,3,2,1])
```
