```python
#L1 and L2 must be permutations of each other
def standard_perm(L1, L2):
    perm = Word(L2).standard_permutation() / Word(L1).standard_permutation()
    # L1[perm[i]] = L2[i]
    #note this is a right action on L1
    return perm
standard_perm([1,2,3], [3,1,2])
```

```python
#p is a numerical partition
#computes schur_lambda(1,1,...,1)
def Schur_evaluate_num(p, d = None):
    if (d is None):
        d = sum(p)
    schur = s(p).expand(d)
    dim = schur([1 for i in range(d)])
    return dim
x = [2,1,1]
Schur_evaluate_num(x,6)
```

```python
#p is a sylbolic partition
def irrep_Sym_symb(p):
    n = len(p)
    dim = 1
    for i in range(n):
        for j in range(i+1,n):
            dim = dim * ((p[i]-p[j])/(j-i) + 1)
    return dim

vars = []
for i in range(4):
    vars.append(var('x{}'.format(i)))
vars
a = irrep_Sym_symb(vars) #schur_{x0,x1,x2,...}(1,1,...,1)
print(a)
a.subs(x0 == 2, x1 == 1, x2 == 1, x3 == 0)
```

```python
def conjugate(p):
    x = []
    l = len(p)
    for i in range(1,p[0]+1):
        s = 0
        for j in range(l):
            if (p[j] >= i):
                s += 1
            else:
                break
        x.append(s)
    return x
conjugate([4,2,1])
```

```python
def n(p):
    l = len(p)
    s = 0
    for i in range(l):
        s += i * p[i]
    return s
```

```python
def content_poly(p):
    x = var('x')
    l = len(p)
    poly = 1
    for i in range(l):
        for j in range(p[i]):
            poly *= (x+j-i)
    return poly
content_poly([2,1]).subs(x==2)
```

```python
#schur_lambda(1,1,1,....,1), d many 1's
def schur_at_one(p, d = var('d')):
    l = len(p)
    s = content_poly(p).subs(x==d)
    q = conjugate(p)
    for i in range(l):
        for j in range(p[i]):
            #since i,j are indexed from 0, replace i,j by i+1,j+1 respectively
            arm = p[i] - j - 1
            leg = q[j] - i - 1
            s = s / (arm + leg + 1)
    return s
schur_at_one([2,1,1])
```
